# Introdução à redução de dimensionalidade (Feature Selection)

```python
import pandas as pd

data = pd.read_csv("https://raw.githubusercontent.com/alura-cursos/reducao-dimensionalidade/master/data-set/exames.csv")

data
```

```python
data['diagnostico'].unique()
```

```python
from sklearn.model_selection import train_test_split
from numpy import random

SEED = 123321
random.seed(SEED)

valores_exames = data.drop(columns=['diagnostico','id'])
diagnostico = data.diagnostico

treino_x, teste_x, treino_y, teste_y = train_test_split(valores_exames, diagnostico, 
                                                        train_size=0.8, random_state=SEED)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100)
model.fit(treino_x, treino_y)
print(model.score(teste_x, teste_y))
```

## Utilizando Random Forest para todas as dimensões

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from numpy import random

SEED = 123321
random.seed(SEED)

valores_exames = data.drop(columns=['diagnostico','id'])
diagnostico = data.diagnostico
valores_exames_v1 = valores_exames.drop(columns='exame_33') # retirando coluna com valores NaN

treino_x, teste_x, treino_y, teste_y = train_test_split(valores_exames_v1, diagnostico, 
                                                        train_size=0.7, random_state=SEED)

model = RandomForestClassifier(n_estimators=100)
model.fit(treino_x, treino_y)
print("A acurácia foi de %.2f%%" % (model.score(teste_x, teste_y) * 100))
```

### Testando com baseline (dummyclassifier)

```python
from sklearn.dummy import DummyClassifier

SEED = 123321
random.seed(SEED)

classificador_bobo = DummyClassifier(strategy='most_frequent')

classificador_bobo.fit(treino_x, treino_y)
print("A acurácia foi de %.2f%%" % (classificador_bobo.score(teste_x, teste_y) * 100))
```

### Plotando visualizações

```python
dados_plot = pd.concat([diagnostico, valores_exames_v1.iloc[:,0:10]],axis=1)
dados_plot = pd.melt(dados_plot, id_vars='diagnostico',var_name='exames',value_name='valores')

dados_plot.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))
sns.violinplot(x='exames',y='valores',hue='diagnostico',data=dados_plot)

plt.xticks(rotation=90)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

padronizador = StandardScaler()
padronizador.fit(valores_exames_v1)
valores_exames_v2 = padronizador.transform(valores_exames_v1)
valores_exames_v2 = pd.DataFrame(data = valores_exames_v2,
                                 columns = valores_exames_v1.keys())

valores_exames_v2
```

```python
dados_plot = pd.concat([diagnostico, valores_exames_v2.iloc[:,0:10]],axis=1)
dados_plot = pd.melt(dados_plot, id_vars='diagnostico',var_name='exames',value_name='valores')

plt.figure(figsize=(15,10))
sns.violinplot(x='exames',y='valores',
               hue='diagnostico',
               data=dados_plot,
               split = 2)

plt.xticks(rotation=90)
```

```python
def grafico_violino(valores, inicio, fim):
  dados_plot = pd.concat([diagnostico, valores.iloc[:,inicio:fim]],axis=1)
  dados_plot = pd.melt(dados_plot, id_vars='diagnostico',var_name='exames',value_name='valores')

  plt.figure(figsize=(15,10))
  sns.violinplot(x='exames',y='valores',
               hue='diagnostico',
               data=dados_plot,
               split = 2)

  plt.xticks(rotation=90)

grafico_violino(valores_exames_v2,10,21)
```

```python
grafico_violino(valores_exames_v2,22,32)
```

```python
valores_exames_v3 = valores_exames_v2.drop(columns=['exame_29','exame_4'])

def classificar(valores):
  SEED = 123321
  random.seed(SEED)
  treino_x, teste_x, treino_y, teste_y = train_test_split(valores, diagnostico, 
                                                          train_size=0.7, random_state=SEED)

  model = RandomForestClassifier(n_estimators=100)
  model.fit(treino_x, treino_y)
  print("A acurácia foi de %.2f%%" % (model.score(teste_x, teste_y) * 100))
```

```python
classificar(valores_exames_v3)
```

```python
valores_exames_v4 = valores_exames_v2.drop(columns=['exame_29','exame_4','exame_11','exame_16','exame_20','exame_10'])

classificar(valores_exames_v4)
```
